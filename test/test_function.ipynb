```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x for log(0.1x + 1)
x_log_smooth = np.linspace(0, 100, 400)
# Calculate log(0.1x + 1)
y_log_smooth = np.log(x_log_smooth+1)

# Plot the log(0.1x + 1) function
plt.figure(figsize=(8, 6))
plt.plot(x_log_smooth, y_log_smooth, label='log(0.01x + 1)', color='blue')
plt.xlabel('x')
plt.ylabel('log(0.1x + 1)')
plt.title('Plot of log(0.1x + 1)')
plt.legend()
plt.grid(True)
plt.show()
```
